```python
#Task 2.1
def task2_1(filename):
    with open(filename) as f:
        return [int(line) for line in f]
```

```python
#Task 2.2
def task2_2(list_of_integers):
    for i in range(len(list_of_integers) - 1):
        noSwaps = True
        for j in range(len(list_of_integers) - 1 - i):
            if list_of_integers[j] > list_of_integers[j + 1]:
                list_of_integers[j], list_of_integers[j + 1] = list_of_integers[j + 1], list_of_integers[j]
                noSwaps = False
        if noSwaps:
            break
       
    return list_of_integers
```

```python
#Task 2.3
def task2_3(list_of_integers):
    for i in range(len(list_of_integers) - 1):
        if list_of_integers[i] > list_of_integers[i + 1]:
            temp = list_of_integers[i + 1]
            j = i
            while list_of_integers[j] > temp and j > -1:
                list_of_integers[j + 1] = list_of_integers[j]
                j -= 1
        
            list_of_integers[j + 1] = temp
    
    return list_of_integers
```

```python
#Task 2.4
def partition(list_of_integers, l, r):
    ptr = l-1
    for i in range(l, r):
        if list_of_integers[i] < list_of_integers[r]:
            ptr += 1
            list_of_integers[ptr], list_of_integers[i] = list_of_integers[i], list_of_integers[ptr]
    ptr += 1
    list_of_integers[ptr], list_of_integers[r] = list_of_integers[r], list_of_integers[ptr]
    return ptr

def quicksort(list_of_integers, l, r):
    if l > r:
        return
    else:
        partitionIndex = partition(list_of_integers, l, r)
        quicksort(list_of_integers, l, partitionIndex - 1)
        quicksort(list_of_integers, partitionIndex + 1, r)

def task2_4(list_of_integers):
    l = 0
    r = len(list_of_integers) - 1
    quicksort(list_of_integers, l, r)
    return list_of_integers
```

```python
#Task 2.5
def merge(arr1, arr2):
    ptr1 = ptr2 = 0
    merged = []
    while ptr1 < len(arr1) and ptr2 < len(arr2):
        if arr1[ptr1] < arr2[ptr2]:
            merged.append(arr1[ptr1])
            ptr1 += 1
        else:
            merged.append(arr2[ptr2])
            ptr2 += 1
    
    while(ptr1 < len(arr1)):
        merged.append(arr1[ptr1])
        ptr1 += 1
        
    while(ptr2 < len(arr2)):
        merged.append(arr2[ptr2])
        ptr2 += 1
    
    return merged

def task2_5(list_of_integers):
    if len(list_of_integers) <= 1:
        return list_of_integers
    else:
        mid = len(list_of_integers) // 2
        left = task2_5(list_of_integers[:mid])
        right = task2_5(list_of_integers[mid:])
        return merge(left, right)
```
