```python
#Task 3.1
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None

class DataStructure:
    def __init__(self):
        self.headPtr = self.tailPtr = None
        
    def display(self):
        current = self.headPtr
        if current is None:
            print("Data Structure is Empty")
        else:
            while(current is not None):
                print(current.data)
                current = current.next
```

```python
#Task 3.2
class Stack(DataStructure):
    def __init__(self):
        super().__init__()
        
    def insert(self, value):
        newNode = Node(value)
        if self.headPtr is None:
            self.headPtr = self.tailPtr = newNode
            
        else:
            newNode.prev = self.tailPtr
            self.tailPtr = newNode
            
    def delete(self):
        if self.tailPtr is None:
            print("Stack is empty.")
        
        else:
            toDelete = self.tailPtr.data
            self.tailPtr = self.tailPtr.prev
            
            if self.tailPtr is None:
                self.headPtr = None
            else:
                self.tailPtr.next = None
            
            return toDelete
        
    def display(self):
        current = self.tailPtr
        if current is None:
            print("Data Structure is Empty")
        else:
            while(current is not None):
                print(current.data)
                current = current.prev
```

```python
#Task 3.3
class Queue(DataStructure):
    def __init__(self):
        super().__init__()
        
    def insert(self, value):
        newNode = Node(value)
        if self.headPtr is None:
            self.headPtr = self.tailPtr = newNode
            
        else:
            self.tailPtr.next = newNode
            self.tailPtr = newNode
            
    def delete(self):
        if self.headPtr is None:
            print("Stack is empty.")
        
        else:
            toDelete = self.headPtr.data
            self.headPtr = self.headPtr.next
            
            if self.headPtr is None:
                self.tailPtr = None
            
            return toDelete
        
    def display(self):
        current = self.headPtr
        if current is None:
            print("Data Structure is Empty")
        else:
            while(current is not None):
                print(current.data)
                current = current.next
```

```python
#Task 3.4
stack = Stack()
with open("TASK3stack.txt") as f:
    for row in f:
        stack.insert(row.strip())

print("Current content of stack:")        
stack.display()
print(f"{stack.delete()} removed")
print(f"{stack.delete()} removed")
print("Current content of Stack:")        
stack.display()

queue = Queue()
with open("TASK3queue.txt") as f:
    for row in f:
        queue.insert(row.strip())

print("Current content of queue:")        
queue.display()
print(f"{queue.delete()} removed")
print(f"{queue.delete()} removed")
print("Current content of queue:")        
queue.display()
```
