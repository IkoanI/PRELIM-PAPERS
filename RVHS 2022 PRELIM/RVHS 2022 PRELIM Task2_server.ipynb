```python
from socket import socket
import random
server = socket()
server.bind(("127.0.0.1", 12345))
server.listen()
newSocket, addr = server.accept()
newSocket.sendall(b"Client connected to server.\n")
while(True):
    message = ""
    while('\n' not in message):
        message += newSocket.recv(1024).decode()
    
    if message == 'START\n':
        answer = ""
        attempts = 5
        history = []
        for i in range(4):
            answer += str(random.randint(1,6))
        
        response = f"New game starts, with {attempts} attempts.\n"
        newSocket.sendall(response.encode())
    
    elif message[:5] == "GUESS":
        guess = message[6:10]
        if guess == answer:
            response = f"You have WON. The code is {answer}.\n"
            newSocket.sendall(response.encode())
            
        elif attempts == 0:
            response = f"You have run out of attempts. The code is {answer}.\n"
            newSocket.sendall(response.encode())
            
        else:
            attempts -= 1
            result = ""
            guessCopy = list(guess)
            answerCopy = list(answer)
            for index, digit in enumerate(guessCopy):
                if digit == answerCopy[index]:
                    result += 'B'
                    answerCopy[index] = None
                    guessCopy[index] = None
            
            for digit1 in guessCopy:
                if digit1 is None:
                    continue
                for index2, digit2 in enumerate(answerCopy):
                    if digit2 is None:
                        continue
                    elif digit1 == digit2:
                        result += 'W'
                        answerCopy[index] = None
            
            if result == "":
                result = "----"
            history.append(guess)
            history.append(result)
            response = f"Guess Result: {result}\n"
            newSocket.sendall(response.encode())
    
    elif message == "HISTORY\n":
        response = f"HISTORY,{attempts},"
        for index, item in enumerate(history):
            if index == len(history) - 1:
                response += f"{item}\n"
            else:
                response += f"{item},"
        
        newSocket.sendall(response.encode())
        
    elif message == "QUIT\n":
        newSocket.sendall(b"Game program ends.\n")
        break

newSocket.close()
server.close()
```
