```python
#Task 2.1
def validate_guesscode(guesscode):
    if len(guesscode) != 4:
        return False
    
    if not guesscode.isdigit():
        return False
    
    for digit in guesscode:
        if int(digit) < 1 or int(digit) > 6:
            return False
        
    return True

print(validate_guesscode("1122"))
print(validate_guesscode("2237"))
```

```python
#Task 2.2
from socket import socket
def menu():
    print("Client initiates connection to server.")
    client = socket()
    client.connect(("127.0.0.1", 12345))
    while(True):
        message = ""
        while('\n' not in message):
            message += client.recv(1024).decode()
        print(message)
        if message == "Game program ends.\n":
            break
        
        print("Master Mind Game Menu")
        print("1) Just start a new game")
        print("2) Make a guess code")
        print("3) Display history")
        print("4) Quit")
        
        option = input("Type an option: ")
        while(option not in ['1', '2', '3', '4']):
            print("Invalid option.")
            option = input("Type an option: ")

        if option == '1':
            client.sendall(b'START\n')

        elif option == '2':
            while(True):
                guess = input("Make a guess: ")
                if validate_guesscode(guess):
                    break
                print("Invalid Guess.")

            message = f"GUESS,{guess}\n"
            client.sendall(message.encode())

        elif option == '3':
            client.sendall(b'HISTORY\n')

        elif option == '4':
            client.sendall(b'QUIT\n')

    client.close()
    
menu()
```
