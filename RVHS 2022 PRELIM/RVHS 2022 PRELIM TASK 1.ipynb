```python
#Task 1.1
class Node:
    def __init__(self, data, left = None, right = None):
        self.data = data
        self.left = left
        self.right = right
        
    
class BST:
    def __init__(self):
        self.root = None
        
    def insert(self, node):
        if self.root is None:
            self.root = node
        else:
            current = self.root
            while(True):
                if node.data < current.data:
                    if current.left is None:
                        current.left = node
                        break
                    else:
                        current = current.left
                else:
                    if current.right is None:
                        current.right = node
                        break
                    else:
                        current = current.right
                
    
    def preOrder(self, root):
        if root is None:
            return ''
        
        else:
            return root.data + self.preOrder(root.left) + self.preOrder(root.right)
    
    
    def inOrder(self, root):
        if root is None:
            return ''
        
        else:
            return self.inOrder(root.left) + root.data + self.inOrder(root.right)

            
    def postOrder(self, root):
        if root is None:
            return ''
        
        else:
            return self.postOrder(root.left) + self.postOrder(root.right) + root.data
            
    def countLeaves(self, root, leaves):
        if root.left is None and root.right is None:
            return 1
        
        leftLeaves = 0
        rightLeaves = 0
        if root.left:
            leftLeaves += self.countLeaves(root.left, leaves) 
        if root.right:
            rightLeaves += self.countLeaves(root.right, leaves)
            
        return leftLeaves + rightLeaves


def task1_1():
    with open("Task1_1.txt") as f:
        for index, line in enumerate(f):
            newBST = BST()
            for char in line.strip():
                newNode = Node(char.strip())
                newBST.insert(newNode)
            
            noLeaves = newBST.countLeaves(newBST.root, 0)
            preOrder = newBST.preOrder(newBST.root)
            inOrder = newBST.inOrder(newBST.root)
            postOrder = newBST.postOrder(newBST.root)
            print(index + 1, noLeaves, line.strip(), preOrder, inOrder, postOrder)
            
task1_1()         
```

```python
def task1_2():
    with open("Task1_2.txt") as f:
        wrongLine = []
        for line in f:
            lineNo, order, given, ans, mark = line.strip().split(',')
            newBST = BST()
            for char in given:
                newNode = Node(char)
                newBST.insert(newNode)
            
            if order == "preorder":
                if newBST.preOrder(newBST.root) != ans and mark != "Incorrect":
                    wrongLine.append(lineNo)
            
            elif order == "inorder":
                if newBST.inOrder(newBST.root) != ans and mark != "Incorrect":
                    wrongLine.append(lineNo)
            
            elif order == "postorder":
                if newBST.postOrder(newBST.root) != ans and mark != "Incorrect":
                    wrongLine.append(lineNo)
                    
        return wrongLine
            
task1_2()
```
