```python
# Task 2.1
import datetime
class LibraryItem:
    def __init__(self, item_id, title, loaned_by = '', due_date = "YYYY-MM-DD"):
        self.item_id = item_id
        self.title = title
        self.loaned_by = loaned_by
        self.due_date = due_date
        
    def is_on_loan(self):
        if self.loaned_by == '':
            return False
        return True
    
    def return_item(self):
        self.loaned_by = ''
        
    def loan_to(self, user_id):
        if self.is_on_loan():
            return False
        else:
            self.loaned_by = user_id
            today = datetime.date.today()
            newDueDate = today + datetime.timedelta(weeks=3)
            self.due_date = newDueDate.strftime('%d-%m-%Y')
            return True
    
    def print_details(self):
        print(f"Title: {self.title}")
        if self.is_on_loan():
            print(f"Loaned by: {self.loaned_by}")
        print(f"Due date: {self.due_date}")
```

```python
# Task 2.2
class Book(LibraryItem):
    def __init__(self, item_id, title, author, category, reserved_by = ''):
        super().__init__(item_id, title)
        self.author = author
        self.category = category
        self.reserved_by = reserved_by
    
    def is_reserved(self):
        if self.reserved_by == '':
            return False
        return True
    
    def release(self):
        self.is_reserved = ''
    
    def reserve_for(self, user_id):
        if self.is_reserved():
            return False
        else:
            self.reserved_by = user_id
            return True
    
    def print_details(self):
        print(f"Title: {self.author}")
        print(f"Author: {self.title}")
        print(f"Category: {self.category}")
        if self.is_on_loan():
            print(f"Loaned by: {self.loaned_by}")
        if self.is_reserved():
            print(f"Reserved by: {self.reserved_by}")
        print(f"Due date: {self.due_date}\n")
        
class CompactDisc(LibraryItem):
    def __init__(self, item_id, title, artist, genre):
        super().__init__(item_id, title)
        self.artist = artist
        self.genre = genre
    
    def print_details(self):
        print(f"Title: {self.title}")
        print(f"Artist: {self.artist}")
        print(f"Genre: {self.genre}")
        if self.is_on_loan():
            print(f"Loaned by: {self.loaned_by}")
        print(f"Due date: {self.due_date}\n")
```

```python
# Task 2.3
with open("libraryitems.txt") as f:
    data = f.readlines()
    items = [item.strip().split(',') for item in data]

libraryItems = []
for item in items:
    if item[0][0] == 'B':
        newBook = Book(item_id = item[0], title = item[1], author = item[2], category = item[3])
        newBook.print_details()
        libraryItems.append(newBook)
    
    elif item[0][0] == 'C':
        newDisc = CompactDisc(item_id = item[0], title = item[1], artist = item[2], genre = item[3])
        newDisc.print_details()
        libraryItems.append(newDisc)
```

```python
# Task 2.4
for item in libraryItems:
    item.loan_to("S1111111H")
    
for item in libraryItems:
    if item.item_id[0] == 'B':
        item.reserve_for("S1222222D")
        
for item in libraryItems:
    item.print_details()
```
