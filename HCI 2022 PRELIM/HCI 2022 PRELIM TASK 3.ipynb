```python
# Task 3.1
class Node:
    def __init__(self, data, ptr = None):
        self.data = data
        self.ptr = ptr

class LinkedList:
    def __init__(self):
        self.head = None
        self.size = 0
        
    def insert(self, word, p):
        if p < 1:
            print("Invalid Element")
        
        else:
            newNode = Node(word)
            
            if self.size == 0:
                self.head = newNode
                self.size += 1
                
            elif p == 1:
                newNode.ptr = self.head
                self.head = newNode
            
            else:
                currentNode = self.head
                nextNode = currentNode.ptr
                currentElement = 1
                while p != currentElement + 1 and nextNode is not None: 
                    currentNode = nextNode
                    nextNode = currentNode.ptr
                    currentElement += 1

                if p == currentElement + 1:
                    newNode.ptr = nextNode
                    currentNode.ptr = newNode
                    self.size += 1
                else:
                    currentNode.ptr = newNode
                    self.size += 1
        
    def delete(self, p):
        if self.size == 0:
            print("Linked list is empty!")
        
        else:
            if self.size == 1:
                self.head = None
                self.size -= 1
                
            elif p == 1:
                self.head = self.head.ptr
            
            else:
                previousNode = None
                currentNode = self.head
                currentElement = 1
                while currentElement < p and currentNode is not None:
                    previousNode = currentNode
                    currentNode = currentNode.ptr
                    currentElement += 1
                
                if currentElement == p:
                    previousNode.ptr = currentNode.ptr
                    self.size -= 1
                else:
                    previousNode.ptr = None
                    self.size -= 1
                    
    def search(self, word):
        currentNode = self.head
        found = False
        
        while(not found and currentNode is not None):
            if currentNode.data == word:
                found = True
            else:
                currentNode = currentNode.ptr
        
        return found
    
    def to_String(self):
        if self.size == 0:
            print("Linked list is empty!")
            
        else:
            toPrint = ''
            currentNode = self.head
            
            while(currentNode is not None):
                toPrint += f"{currentNode.data}, "
                currentNode = currentNode.ptr
            
            print(toPrint.strip().strip(','))
            
```

```python
ll = LinkedList()
ll.to_String()
ll.delete(1)
ll.insert("apple", 1)
ll.to_String()
ll.insert("carrot", 3)
ll.to_String()
ll.delete(1)
ll.to_String()
ll.insert("apple", 1)
ll.to_String()
ll.insert("banana", 3)
ll.to_String()
ll.delete(2)
ll.to_String()
ll.insert("carrot", 2)
ll.to_String()
print(ll.search("banana"))
print(ll.search("orange"))
```
